```python
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

file = "../feminicidios_ceara_2018.csv"
data = pd.read_csv(file, sep=';')
```

```python
def plot_line_tot_month(data):    
    label_x = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 
               'Out', 'Nov', 'Dez']
    y = data['Total']

    output_notebook()
    
    p = figure(title="Feminicídios no Ceará - 2018 (dados do Instituto Maria da Penha)",
               x_range=label_x, y_range=(0,60))
    p.line(label_x, y, line_width=3, color="red")
    
    show(p)
```

```python
def plot_bar_age(data):
    young = sum(data['15-24 anos'])
    adult = sum(data['Mais de 25 anos'])

    output_notebook()
    
    label_x = ["< 25 anos", ">= 25 anos"]
    y = [young, adult]
    
    p = figure(title="Feminicídios no Ceará por faixa etária - 2018 (dados do Instituto Maria da Penha)",
               x_range=label_x, y_range=(0, 250))
    p.vbar(x=label_x, width=0.5, bottom=0, top=y, color=["firebrick","purple"])
    
    show(p)
```

```python
plot_line_tot_month(data)
```

```python
plot_bar_age(data)
```
